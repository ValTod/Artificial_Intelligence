```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import j1

x = np.linspace(0, 500, 1000)
y = j1(x)

plt.plot(x, y)
plt.title("First-Order Bessel Function of the First Kind: $J_1(x)$")
plt.xlabel("x")
plt.ylabel("$J_1(x)$")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
lambda_ = 1.0  # Wavelength (arbitrary units)
I0 = 1.0      # Maximum current (arbitrary units)
l = 0.5       # Length of the patch (arbitrary units)
k = 2 * np.pi / lambda_  # Wave number

# E-plane (phi = 90 degrees)
theta = np.linspace(0.0001, 2 * np.pi, 360)
E_theta = (1 / (1j * lambda_)) * (I0 / 2) * np.sin(theta) * (np.sin(k * l * np.cos(theta / 2)) / np.cos(theta / 2))

# H-plane (theta = 90 degrees)
phi = np.linspace(0.0001, 2 * np.pi, 360)
H_phi = (1 / (1j * lambda_)) * (I0 / 2) * np.sin(phi) * (np.sin(k * l * np.sin(phi / 2)) / np.sin(phi / 2))

# Convert to absolute values for plotting
E_theta_magnitude = np.abs(E_theta)
H_phi_magnitude = np.abs(H_phi)

# Plotting E-plane radiation pattern
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1, projection='polar')
plt.plot(theta, E_theta_magnitude)
plt.title('E-plane Radiation Pattern ($\\phi = 90^\\circ$)')
plt.xlabel('$\\theta$ (radians)')
plt.ylabel('$|E(\\theta)|$')

# Plotting H-plane radiation pattern
plt.subplot(1, 2, 2, projection='polar')
plt.plot(phi, H_phi_magnitude)
plt.title('H-plane Radiation Pattern ($\\theta = 90^\\circ$)')
plt.xlabel('$\\phi$ (radians)')
plt.ylabel('$|H(\\phi)|$')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
lambda_ = 1.0  # Wavelength (arbitrary units)
k = 2 * np.pi / lambda_

# Dipole antenna radiation pattern
theta = np.linspace(0, 2 * np.pi, 360)
E_dipole = np.abs(np.sin(theta))

# Monopole antenna radiation pattern
E_monopole = np.abs(np.sin(theta))

# Yagi antenna radiation pattern (simplified)
E_yagi = np.abs(np.cos(theta) ** 1.5)

# Parabolic reflector radiation pattern (simplified)
E_parabolic = np.abs(np.sin(theta)) * np.exp(-theta**2)

# Patch antenna radiation pattern (simplified)
E_patch = np.abs(np.cos(theta) ** 2)

# Plotting
plt.figure(figsize=(15, 10))

# Dipole
plt.subplot(2, 3, 1, projection='polar')
plt.plot(theta, E_dipole)
plt.title('Dipole Antenna')
plt.grid(True)

# Monopole
plt.subplot(2, 3, 2, projection='polar')
plt.plot(theta, E_monopole)
plt.title('Monopole Antenna')
plt.grid(True)

# Yagi
plt.subplot(2, 3, 3, projection='polar')
plt.plot(theta, E_yagi)
plt.title('Yagi Antenna')
plt.grid(True)

# Parabolic Reflector
plt.subplot(2, 3, 4, projection='polar')
plt.plot(theta, E_parabolic)
plt.title('Parabolic Reflector Antenna')
plt.grid(True)

# Patch
plt.subplot(2, 3, 5, projection='polar')
plt.plot(theta, E_patch)
plt.title('Patch Antenna')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
lambda_ = 1.0  # Wavelength (arbitrary units)
k = 2 * np.pi / lambda_

# Angles for radiation patterns
theta = np.linspace(0, 2 * np.pi, 360)

# Dipole antenna radiation pattern
E_dipole = np.abs(np.sin(theta))

# Monopole antenna radiation pattern
E_monopole = np.abs(np.sin(theta))

# Yagi antenna radiation pattern (simplified)
E_yagi = np.abs(np.cos(theta) ** 1.5)

# Parabolic reflector radiation pattern (simplified)
E_parabolic = np.abs(np.sin(theta)) * np.exp(-theta**2 / (2 * (np.pi/8)**2))

# Patch antenna radiation pattern (simplified)
E_patch = np.abs(np.cos(theta) ** 2)

# Plotting all patterns in one plot
plt.figure(figsize=(10, 8))

plt.subplot(111, projection='polar')
plt.plot(theta, E_dipole, label='Dipole Antenna', color='blue')
plt.plot(theta, E_monopole, label='Monopole Antenna', color='green')
plt.plot(theta, E_yagi, label='Yagi Antenna', color='red')
plt.plot(theta, E_parabolic, label='Parabolic Reflector Antenna', color='purple')
plt.plot(theta, E_patch, label='Patch Antenna', color='orange')

plt.title('Radiation Patterns of Different Antennas')
plt.legend(loc='upper right')
plt.grid(True)

plt.show()
```
